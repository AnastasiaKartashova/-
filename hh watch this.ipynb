```python
params = {}
vacancies = []
for page in range(0, 42):
  params['page'] = page
  req = BeautifulSoup(requests.get(url, params=params, headers = headers).text)
  v = req.find_all('a', {'data-qa': 'serp-item__title'}, href = True)
  vacancies.extend(v)
  sleep(randint(3, 8))
```

```python
vacancies = vacancies[2:]
```

```python
links = [link['href'].split('https://hh.ru/')[1] for link in vacancies]
```

```python
links = [link.split('?')[0] for link in links]
```

```python
exp = []
descr = []
requirments = []
for l in links:
  reqs = []
  req = BeautifulSoup(requests.get(l, headers=headers).text)
  e = req.find('span', {"data-qa":"vacancy-experience"}).text
  d = req.find('div', {'data-qa':'vacancy-description'}).text
  r = req.find_all('div', class_='bloko-tag bloko-tag_inline')
  for j in range(len(r)):
    requirment = r[j].text
    reqs.append(requirment)
  exp.append(e)
  descr.append(d)
  requirments.append(reqs)
  sleep(randint(3, 8))
```

```python
exp1 = []
descr1 = []
requirments1 = []
for l in links[156:]:
  reqs = []
  req = BeautifulSoup(requests.get(l, headers=headers).text)
  e = req.find('span', {"data-qa":"vacancy-experience"}).text
  d = req.find('div', {'data-qa':'vacancy-description'}).text
  r = req.find_all('div', class_='bloko-tag bloko-tag_inline')
  for j in range(len(r)):
    requirment = r[j].text
    reqs.append(requirment)
  exp1.append(e)
  descr1.append(d)
  requirments1.append(reqs)
  sleep(randint(3, 8))
```

```python
exp2 = []
descr2 = []
requirments2 = []
for l in links[233:]:
  reqs = []
  req = BeautifulSoup(requests.get(l, headers=headers).text)
  e = req.find('span', {"data-qa":"vacancy-experience"}).text
  d = req.find('div', {'data-qa':'vacancy-description'}).text
  r = req.find_all('div', class_='bloko-tag bloko-tag_inline')
  for j in range(len(r)):
    requirment = r[j].text
    reqs.append(requirment)
  exp2.append(e)
  descr2.append(d)
  requirments2.append(reqs)
  sleep(randint(3, 8))
```

```python
print(len(exp)+len(exp1)+len(exp2))
print(len(descr)+len(descr1)+len(descr2))
print(len(requirments)+len(requirments1)+len(requirments2))
```
